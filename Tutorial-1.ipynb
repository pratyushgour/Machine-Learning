```python
from numpy import random
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.random.normal(5, 1, 10)
```

```python
m = np.linspace(0,10,50)
```

```python
l = []
mean = []
for i in range(len(m)):
    ret = (1/np.sqrt(2*np.pi))*np.exp(-0.5*(x-m[i])**2)
    l.append(np.prod(ret))
    mean.append(m[i])
```

```python
plt.plot(mean, l)
```

```python
logl = np.log10(l)
```

```python
plt.plot(mean, logl)
```
